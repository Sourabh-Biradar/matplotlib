```python
# matplotlib : used to vizualize data ; plot graphs

import matplotlib.pyplot as plt

data = [3,2,4,3]    

plt.plot(data)      # y-axis points ; by default x-axis points = indices ; plt.plot(x,y)

plt.title('Simple graph') # name of graph

plt.ylabel('nums')        # y-axis name
plt.xlabel('index')       # x-axis name

plt.show()                # optinal in jupyter bt recmd to use everywhr
```

```python
# graph - 2 : points colours & lineshape , grid

n = [1,2,3,4]
sq = [1,4,9,16]

plt.plot(n,sq,color='red',marker='o',linestyle=':')              # plt.plot(x,y)  

# plt.plot(n,sq,'ro:')                                           # shorthand : 'ro:' = red circles & dotted line

plt.grid()
plt.show()

# color marker linestyle
# 'mD-'  : magenta diamond solid line
# 'b*:'  : blue star dotted line
# 'ro'   : red circle no line
# 'k^--' : black triangle dashed line
```

```python
# graph - 3 : multiple plots ,line label

yr_2021 = [4,8,7,2,9]
yr_2022 = [5,2,1,8,3]
yr_2023 = [9,2,8,2,7]

plt.plot(yr_2021,'mD-',label='2021')
plt.plot(yr_2022,'k*:',label='2022')
plt.plot(yr_2023,'g^-.',label='2023')

plt.legend()  # displayes small square on graph

plt.show()
```

```python
# linewidth, setp() : set properties 

x1 = [2,4,6,8]
y1 = [3,6,9,12]

x2 = [8,3,5,1]
y2 = [4,8,1,3]

lines = plt.plot(x1,y1,x2,y2)    # plots 2 lines

plt.setp(lines[0],color='grey',linewidth=2.0)         # lines[0] = x1,y1
plt.setp(lines[1],'color','green','linewidth',2.0)    # lines[1] = x2,y2 (MATLAB style)

plt.show()
```

```python
# figure() , subplot()

x1 = [2,4,6,8]
y1 = [3,12,9,2]

x2 = [8,3,5,1]
y2 = [4,8,1,3]

plt.figure(1)
plt.subplot(211)                 # subplot(rows-columns-figure) how many rows
plt.plot(x1,y1,label='211')  
plt.title('subplot 211')

plt.subplot(212)
plt.plot(x2,y2)
plt.title('subplot 212')

plt.tight_layout()           # avoids overlaping
plt.show()
```
